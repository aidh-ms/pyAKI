# Prep Validated Data for Automatic Tests

```python
import pandas as pd

validation = pd.read_excel('tests/data/test_data/test_aki_combined_FA_validated_Analysis.xlsx')
validation.head()
```

```python
validation.info()
```

```python
# beacuse I'm an idiot...
validation.rename(columns={'creatinine_stag_vote': 'creatinine_stage_vote'}, inplace=True)
```

```python
# forward fill stay_id
validation.stay_id = validation.stay_id.ffill()
validation.head()
```

```python
# convert stay_id to int
validation.stay_id = validation.stay_id.astype(int)
```

```python
def define_true_label(row, label):
    if label == "stage":
        human_label = "human_" + label
        machine_label = "machine_" + label
        vote_label = label + "_vote"
    else:
        human_label = "human_" + label + "_stage"
        machine_label = "machine_" + label + "_stage"
        vote_label = label + "_stage" + "_vote"
    if label == "abs_creatinine" or label == "rel_creatinine":
        vote_label = "creatinine_" + "stage" + "_vote"
    
    if row[human_label] == row[machine_label]:
        return row[machine_label]
    elif row[human_label] != row[machine_label]:
        return row[vote_label]
    else:
        raise ValueError("Something went wrong")
    
validation['true_urineoutput_stage'] = validation.apply(lambda row: define_true_label(row, 'urineoutput'), axis=1)
```

```python
validation['true_abs_creatinine_stage'] = validation.apply(lambda row: define_true_label(row, 'abs_creatinine'), axis=1)
validation['true_rel_creatinine_stage'] = validation.apply(lambda row: define_true_label(row, 'rel_creatinine'), axis=1)
validation['true_crrt_stage'] = validation.apply(lambda row: define_true_label(row, 'crrt'), axis=1)
validation['true_stage'] = validation.apply(lambda row: define_true_label(row, 'stage'), axis=1)
```

```python
validation = validation[[
    "stay_id",
    "charttime",
    "urineoutput",
    "true_urineoutput_stage",
    "creat",
    "true_abs_creatinine_stage",
    "true_rel_creatinine_stage",
    "weight",
    "crrt_status",
    "true_crrt_stage",
    "true_stage"
]].copy()
```
